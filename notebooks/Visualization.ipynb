1- seed = referência para resultados reproduzíveis
2- L = eh o z nas formulas de adsorção = comprimento do leito
3- Nz = número de pontos no leito em quantos pedacos a malha foi dividida
4- eps = porosidade do leito
5- rho_b = densidade aparente do adsorvente kg/m^3
6- u = velocidade do fluido m/s
7- D_ax = coeficiente de difusão axial m^2/s
8- kl = coeficiente de transferência de massa 1/s
9- qmax = capacidade máxima de adsorção mol/kg
10 - b = constante de afinidade 1/concentração
11- n = expoente de Heterogeneidade de Freundlich
12- lam_z = condutividade térmica do leito - dispercao termica axial W/mK
13- rho_g = denssidade do gas kg/m^3
14- cp_g = capacidade termica do gas J/kgK
15- cp_s = capacidade termica do solido J/kgK
16- D_col = diametro interno da coluna em metros
17- h_w = Coeficiente de transferência de calor para a parede (W/m²·K)
18- T_wall = Temperatura da parede (K)
19- dH = Calor de adsorção (J/mol, tipicamente negativo)
20- dt = Intervalos de tempo (s)
21- t_end = tempo final da simulacao (s)
22- C_in = Concentração na entrada (mol/m³, etc.)
23- T_in = Temperatura na entrada (K)
24- C_out_final = Concentração do gás na saída da coluna (última célula em z) no tempo final da simulação.
25- q_out_final = Carga adsorvida na região de saída (ou média da coluna) no tempo final.
26- T_out_final = Temperatura na saída da coluna no tempo final.
27- N_ads_final = Quantidade total adsorvida no leito (integral de q ao longo da coluna) no tempo final.
28- C_z* = concentração do gás na posição z_k (k-ésima célula) no tempo final.
29- q_z* = quanto está adsorvido no sólido naquela mesma posição.
30- T_z* = temperatura local naquela posição.
31- Qtot_t* = quantidade total adsorvida no leito no tempo t_*.
